```python
import pandas as pd
```

```python
customers_df = pd.read_csv('C:/Users/Administrator/Documents/trial/customers.csv')
customers_df.head()
```

```python
orders_df = pd.read_csv('C:/Users/Administrator/Documents/trial/orders.csv')
orders_df.head()
```

```python
transactions_df = pd.read_csv('C:/Users/Administrator/Documents/trial/transactions.csv')
```

```python
transactions_df.head()
```

```python
# Filter transactions for high-value transactions
high_value_transactions = transactions_df[transactions_df['sales'] > 150]

# Get a list of order IDs with high-value transactions
high_value_order_ids = high_value_transactions['order_id'].unique()

# Filter orders for the order IDs with high-value transactions
high_value_orders = orders_df[orders_df['order_id'].isin(high_value_order_ids)]

# Get a list of customer IDs who placed high-value orders
high_value_customer_ids = high_value_orders['customer_id'].unique()

# Filter customers for the customer IDs who placed high-value orders
high_value_customers = customers_df[customers_df['customer_id'].isin(high_value_customer_ids)]

# Display the results
print("List of order IDs with high-value transactions:")
print(high_value_order_ids)

print("List of customers who placed orders with high-value transactions:")
print(high_value_customers[['customer_id', 'customer_name']])
```

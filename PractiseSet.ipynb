```python
import pandas as pd
```

```python
customers_df = pd.read_csv('C:/Users/Administrator/Documents/trial/customers.csv')
customers_df.head()
```

```python
orders_df = pd.read_csv('C:/Users/Administrator/Documents/trial/orders.csv')
orders_df.head()
```

```python
transactions_df = pd.read_csv('C:/Users/Administrator/Documents/trial/transactions.csv')
```

```python
transactions_df.head()
```

```python
# Filter transactions for high-value transactions based on the new threshold (sales > 250)
high_value_transactions = transactions_df[transactions_df['sales'] > 250]

# Get a list of order IDs with high-value transactions
high_value_order_ids = high_value_transactions['order_id'].unique()

# Filter orders for the order IDs with high-value transactions
high_value_orders = orders_df[orders_df['order_id'].isin(high_value_order_ids)]

# Get a list of customer IDs who placed high-value orders
high_value_customer_ids = high_value_orders['customer_id'].unique()

# Filter customers for the customer IDs who placed high-value orders
high_value_customers = customers_df[customers_df['customer_id'].isin(high_value_customer_ids)]

# Extract the month and year from the 'order_purchase_date' for grouping
high_value_orders['order_purchase_month'] = high_value_orders['order_purchase_date'].str.split(' ', expand=True)[0]
high_value_orders['order_purchase_year'] = high_value_orders['order_purchase_date'].str.split(' ', expand=True)[1]

# Calculate month and year-wise high-value transactions
high_value_transactions_by_month_year = high_value_orders.groupby(['order_purchase_month', 'order_purchase_year']).size()

# Display the results
print("List of order IDs with high-value transactions:")
print(high_value_order_ids)

print("List of customers who placed orders with high-value transactions:")
for customer_id, customer_name in zip(high_value_customers['customer_id'], high_value_customers['customer_name']):
    print(f"Customer ID: {customer_id}, Customer Name: {customer_name}")

print("Month and Year-wise High-Value Transactions:")
print(high_value_transactions_by_month_year)
```
